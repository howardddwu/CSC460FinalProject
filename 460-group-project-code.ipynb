```python
import pandas as pd
pd.options.mode.chained_assignment = None
pd.options.display.max_columns = 999
import numpy as np
import matplotlib.pyplot as plt
```

### Loading data

```python
# Loading data
df = pd.read_csv('../input/us-accidents/US_Accidents_Dec21_updated.csv')
```

### Cleaning data

```python
df.isnull().sum()
```

```python
print(df.shape)
df.dropna(inplace = True)
print(df.shape)
df.drop(columns = ['ID'], inplace = True)
df.drop(columns = ['Description'], inplace = True)
```

### Check correlation

```python
df.corr()
```

### Below is our section where we will try to build a good model on predicting

```python
from sklearn.preprocessing import StandardScaler
X = df[["Start_Lat", "Start_Lng", "End_Lat", "End_Lng", "Distance(mi)"]]
y = df["Severity"].values
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
# Create Train / Test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

### Function to test the model's performance

```python
from sklearn.metrics import accuracy_score

def printResult(model):
    # Predict and measure the accuracy on the training set.
    predicted_y_train = model.predict(X_train)
    predicted_y_train = predicted_y_train.astype(int)
    train_acc = accuracy_score(y_train, predicted_y_train)
    print(f"The training accuracy is: {train_acc*100:.2f}%")

    # Predict and measure the accuracy on the testing set.
    predicted_y_test = model.predict(X_test)
    predicted_y_test = predicted_y_test.astype(int)
    test_acc = accuracy_score(y_test, predicted_y_test)
    print(f"The testing accuracy is {test_acc*100:.2f}%")
```

#### Linear model
- we don't think linear model is good for such problem, but we still want to test it out

```python
from sklearn.linear_model import LinearRegression

# train the linear model using the trainning dataset
linear_model = LinearRegression().fit(X_train, y_train)
printResult(linear_model)
```

- **As we can see that the results are actually not so good. We will not use this model to perdict if there will an accident**

#### Tree model

```python
from sklearn import tree

tree_model = tree.DecisionTreeClassifier(max_depth=1)
tree_model = tree_model.fit(X_train, y_train)
printResult(tree_model)
```

- **This model is much better than the linear model that we use to predict. Decision trees seem to work very well in such binary classification problem.**

```python
### Let's try different depth of the tree_classifier
for depth in range(2, 11):
    model = tree.DecisionTreeClassifier(max_depth=depth)
    model = model.fit(X_train, y_train)
    print("Depth: ", depth)
    printResult(model)
```

- **As the depth increases, both training accuracy and testing accuracy are getting higher. This is pretty good for our model.**
- **As indicated above, the training accuracy is getting higher as well, and this might cause an overfitting problem**
